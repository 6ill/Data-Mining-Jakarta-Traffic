```python
import pandas as pd
import numpy as np
import json
```

```python
# Read the JSON file
with open('results_traffic_jakarta.json', 'r') as file:
    data = json.load(file)

# Normalize the JSON data
df = pd.json_normalize(data)
```

```python
df
```

```python
df.loc[0]['network.segmentResults'][0].keys()
```

```python
df.loc[0]['network.segmentResults'][0]['segmentProbeCounts'][0].keys()
```

```python
cols  = ['segmentId', 'frc', 'speedLimit', 'streetName', 'distance']
segment_probe_cols = ['timeSet', 'dateRange', 'probeCount']
```

```python
streets_list = []
for obj in df.loc[0]['network.segmentResults']:
    if obj['segmentProbeCounts'][0]['probeCount'] == 0 or obj['frc'] > 5 or ('streetName' not in obj) :
        continue


    street_info = dict()

    for col in cols:
        street_info[col] = obj[col]

    for col in segment_probe_cols:
        street_info[col] = obj['segmentProbeCounts'][0][col]

    streets_list.append(street_info)
    print(obj['streetName'])
```

```python
streets_list
```

```python
cleaned_data = pd.DataFrame(streets_list)
cleaned_data
```
